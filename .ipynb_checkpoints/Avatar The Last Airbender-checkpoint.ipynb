# Avatar: The Last Airbender Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# read data
file_path = 'C:/Users/Onur/Desktop/RR github assignments/Assignment 8/Assignment_8/data.xlsx'
df = pd.read_excel(file_path)

```

Avatar: The Last Airbender carved out its niche in storytelling with an intricate blend of fantasy and adventure. Upon its airing, it didn't take long for the series to capture the hearts of viewers, young and old. The tale of Aang's journey, burdened with the destiny of an Avatar, was set against a backdrop rich with mythology and cultures that felt as tangible as the elements its characters. This depth, coupled with a balanced mix of humor, heart, and heroism, propelled the series beyond its Saturday morning slot into a fixture of early-2000s pop culture.

As fans tuned in, captivated by episodes like "The Boy in the Iceberg" and "The Warriors of Kyoshi," their engagement was quantified through votes and ratings. Consistently high marks across the seasons reflect a quality of content that rarely wavered and often exceeded expectations. The figures offer a statistical echo of the acclaim the series has enjoyed, standing as a metric of its enduring appeal and the legacy of its story.

### Rating Distribution by Season

```python
atla_colors = {1: '#00bfff', 2: '#6b8e23', 3: '#b22222'}

plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='Average Rating', data=df, palette=atla_colors)
plt.title('Distribution of Average Ratings per Season')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

#### Season 1 (Water)
    The ratings for the first season show a broader range, with the median rating sitting around the 8.5 mark. The spread suggests more variability in how viewers rated these episodes, with some being rated as low as around 7.5 and some as high as close to 9.5.
#### Season 2 (Earth)
    This season shows a higher median rating than Season 1, indicating a generally more favorable reception. The interquartile range (IQR) is tighter, which shows that the ratings are more consistent around the median, hovering above 8.5 but rarely going below 8.0.
#### Season 3 (Fire): 
    The third season displays the highest median rating, which is around 9.0, suggesting that this season was very well-received. Although there's an episode or two rated around 8.0 (indicated by the lower "whisker" of the boxplot), the overall ratings are quite high, with the IQR predominantly above 8.5.
    
Season 3 appears to be the most consistently highly-rated season, while Season 1 shows the most variability in ratings. The whiskers of the plots, which indicate the range of the ratings outside the middle 50% (excluding outliers), do not differ dramatically between seasons, suggesting that while there are episodes with lower ratings in each season, they are not extreme outliers.

### Most and Least Voted Episode

```python
# Sort and select the top and bottom 5 episodes based on total votes
top_voted = df.nlargest(5, 'total votes')
bottom_voted = df.nsmallest(5, 'total votes').sort_values('total votes', ascending=False)

# Map the season to the respective color
top_voted_colors = top_voted['season'].map(atla_colors)
bottom_voted_colors = bottom_voted['season'].map(atla_colors)

# Create a new figure with a specific size
plt.figure(figsize=(12, 6))

# Create a bar chart for the top voted episodes
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.bar(top_voted['title'], top_voted['total votes'], color=top_voted_colors)
plt.title('Top 5 Voted Episodes')
plt.xlabel('Episode')
plt.ylabel('Total Votes')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

# Create a bar chart for the least voted episodes
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.bar(bottom_voted['title'], bottom_voted['total votes'], color=bottom_voted_colors)
plt.title('Least 5 Voted Episodes')
plt.xlabel('Episode')
plt.ylabel('Total Votes')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

This graph tells a tale of audience engagement evolving over the show's lifespan. Season 3's pivotal episodes, marked in red, likely a mix of climax hype and the great farewell to a beloved series. In contrast, the episodes from Seasons 2, shown in green, have a steadier, though less pronounced, voting presence. This pattern speaks to a growing attachment with the series, as the emotional stakes ramped up and story arcs peaked, the fans' involvement deepened, culminating for the a passionate finale.

###  Distribution of Rankings by Season

```python
bins = range(int(df['Average Rating'].min()), int(df['Average Rating'].max()) + 2)

# Initialize a base for the stacks
base = np.zeros(len(bins)-1)

# Plot for each season
for season in sorted(df['season'].unique()):
    # Filter the DataFrame for the season and get the histogram data
    season_data = df[df['season'] == season]['Average Rating']
    hist_data, _ = np.histogram(season_data, bins=bins)
    
    # Plot the bars for this season
    plt.bar(bins[:-1], hist_data, bottom=base, color=atla_colors[season], width=1, label=f'Season {season}')
    
    # Update the base for the next season's stack
    base += hist_data

# Set the attributes for the plot
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.xticks(bins) 
plt.legend(title='Season')

# Show the plot
plt.show()
```

The stacked histogram illustrates the shifting audience ratings for "Avatar: The Last Airbender" across its three seasons. It's clear from the bands of colorblue for Season 1, green for Season 2, and red for Season 3that as the series progressed, the frequency of higher ratings increased. Season 1 laid the groundwork with a solid base of ratings in the 7 to 8 range. Season 2 built on this foundation, pushing more into the 8 to 9 territory, suggesting a deepening viewer investment in the series. By Season 3, the ratings frequently touched the 9 to 10 mark, indicating peak viewer satisfaction and a strong series finale. Each color not only represents a chapter in Aang's story but also a step up in the series' narrative arc, with the increasing height of each color band mirroring the ascending quality and emotional stakes of the show.

### Episode Ratings Over Time by Season

```python
plt.figure(figsize=(14, 7))

# Create the line plot with the specified palette
sns.lineplot(data=df, x='episode', y='Average Rating', hue='season', palette=atla_colors, marker='o')

plt.xlabel('Episode Number')
plt.ylabel('Average Rating')

max_episode = df['episode'].max()
plt.xticks(range(1, max_episode + 1)) 

# Display the legend and grid
plt.legend(title='Season')
plt.grid(True)

# Show the plot
plt.show()
```

Season 1 showcases a notable outlier early on, where we see a dip that might correspond to an episode that deviated from viewers' expectations or experimented with narrative style. However, as the season progresses, the ratings find a firmer footing, possibly as the series found its rhythm and viewers began to engage more deeply with the characters and story.

When we turn to the season finales, the graph shows a striking trend: each season's culmination is met with resounding approval, evidenced by significant peaks in ratings. These finales, marked by pivotal plot resolutions and climactic events, seem to have delivered exactly what the audience craved. Particularly for Season 3, we witness not just high ratings for the finale but a steady, ascending line leading up to it, indicating a consistent build-up of viewer satisfaction and engagement. This upward trajectory suggests a series successfully raising the stakes and deepening its narrative impact, culminating in a finale that stands as a high point of the series, reflected in its peak ratings.
